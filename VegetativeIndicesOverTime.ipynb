```python
#import necessary libraries

import pandas as pd
import math as math
```

```python
#load in csv of all our data
Data = pd.read_csv(r'C:\Users\Brent\Desktop\ClassNotes\CSS844\Project1\VegetativeIndices\vegetation_indices_results_2.csv')
```

```python
Accumulated_GDD = Data['Accumulated.GDD'][:5] #makes new GDD series
Data = Data.drop('Accumulated.GDD', axis=1) #removes GDD from data
Data_VI = Data.drop('id', axis=1) #makes new Dataframe just for making New_Data
Data_VI = Data_VI.drop('Date', axis=1)
```

```python
#New empty panda to write data into
New_Data = pd.DataFrame()
New_Data['id'] = []
New_Data['id'] = Data['id'].unique() #Makes a row for each ID
for i in range(len(Data_VI.columns)): #Makes a column for each VI_Date
    for j in range(0,5):
        new_column = f"{Data_VI.columns[i]}_{Data['Date'][j]}"
        New_Data[new_column] = None
print(New_Data)
```

```python
#Reorganizing data into the new dataframe
for i in range(len(New_Data.columns)-1): #Iterates through each column
    temp_series = [] #sets up empty series
    for j in range(len(New_Data)): #iterates through each row of that column
        next_value = Data.loc[(j*5)+(i % 5)][math.floor(i/5)+2] #chooses the next value

        temp_series.append(next_value) #adds the value to the series
    New_Data[New_Data.columns[i+1]] = temp_series #replaces column with the series


New_Data.head()
```

```python
#Looking at averages of each VI at each date
Averages = New_Data.iloc[:, 1:].mean()
display(Averages)
display(Accumulated_GDD)
```
